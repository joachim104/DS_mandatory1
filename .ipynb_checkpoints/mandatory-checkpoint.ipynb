```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

```python
pd.options.display.max_rows=3000


df = pd.read_csv('data.csv')

s = "Q"
columns = []
for i in range(1, 29):
    columns.append(s + str(i))
    
df[columns] = df[columns].replace([1, 2, 3, 4, 5], [-2, -1, 0, 1, 2])
df["sum"] = df[columns].sum(axis=1).div(28)
df2 = df[df["Q1"] != df["sum"]]
#df2


attendance = df["attendance"].sum()

float_att = float(attendance)

courseQ = ['Q11', 'Q4', 'Q9','Q10',]

instrQ = ['Q15','Q16', 'Q13', 'Q19']

misc = ['attendance', 'nb.repeat', 'difficulty']

mergedQ = courseQ + instrQ + misc

table = pd.pivot_table(df2, values=mergedQ, 
                    index=['instr', 'class'],
                    aggfunc=np.mean)

table['Course question sum'] = table[courseQ].sum(axis=1).div(4)
table['Instructor question sum'] = table[instrQ].sum(axis=1).div(4)
table['Difference'] = (table['Course question sum'] - table['Instructor question sum'])


table
# mergedQ

```
